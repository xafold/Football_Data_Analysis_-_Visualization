```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Patch
```

```python
df = pd.read_csv(r'C:\Users\LENOVO\Merged_df.csv')
```

```python
df
```

```python
df1 = df[['Squad','LgRk','Fans_Attendance_Percentage']].copy()
```

```python
sns.set_style('whitegrid')
```

```python
df1
```

```python
top_6=df1.nsmallest(10, 'LgRk').drop(index=[2, 5, 6, 8]).reset_index(drop=True)
```

```python
top_6
```

```python
df2=pd.merge(df1, top_6, on='Squad', how='left')
```

```python
df2
```

```python
df2['top_6'] = np.where(df2['LgRk_y'].notnull(), 'top_6', 'others')
```

```python
df2
```

```python
df2 = df2.dropna(axis=1)
```

```python
df2
```

```python
df2=df2.sort_values(by=['Fans_Attendance_Percentage_x','Squad'],ascending=[False, True])
```

```python
df2
```

```python
df2.reset_index(drop=True)
```

```python
df2=df2.reset_index(drop=True)
```

```python
print(df2.columns)
```

```python
plt.figure(figsize=(8, 6))
palette = ['#1750AC' if x in df2[df2['top_6']=='top_6'].Squad.values else '#D3D3D3' for x in df2.Squad]
ax = sns.barplot(y='Squad',x='Fans_Attendance_Percentage_x',data=df2,
            palette=['#0096FF' if x in df2[df2['top_6']=='top_6'].Squad.values else '#E5E4E2' for x in df2.Squad], 
            edgecolor='black', linewidth=0.2,alpha=0.9)

xlim = ax.get_xlim()

for i, rect in enumerate(ax.patches):
    ax.text(xlim[1]*0.5, rect.get_y() + rect.get_height()/2, 
            str(round(df2.Fans_Attendance_Percentage_x[i], 2)) + '%', ha='center', va='center', rotation=0)

plt.ylabel('Clubs')
plt.xlabel('Stadium Fillout (%)')

top_6 = Patch(color='#0096FF', label='Top 6 clubs')
others = Patch(color='#E5E4E2', label='Other clubs')
plt.legend(handles=[top_6, others])
plt.legend(handles=[top_6, others],loc='upper left', bbox_to_anchor=(1, 1))
sns.set_style("ticks")
sns.despine(right=True, top=True)
plt.savefig("myplot4.png", bbox_inches='tight',dpi=1000)


```
